```python
df_suburbs=df_plants['VIC_LOCA_2'].drop_duplicates().reset_index()
```

```python
del df_suburbs['index']
```

```python
df_suburbs.head()
```

```python
df_suburbs=df_suburbs.dropna()
df_suburbs.shape
```

```python
df_suburb_data = pd.read_csv ('australian_postcodes.csv')
df_suburb_data.head()
```

```python
df_suburb_data_subset=df_suburb_data[['Locality','Long','Lat']]
df_suburb_data_subset.head()
```

```python
df_suburb_data_subset = df_suburb_data_subset.drop_duplicates(subset=['Locality'])
df_suburb_data_subset.head()
```

```python
suburb_coordinates=pd.merge(df_suburbs, df_suburb_data_subset, how='left',left_on='VIC_LOCA_2', right_on='Locality')
del suburb_coordinates['Locality']
suburb_coordinates.head()
```

```python
suburb_coordinates=suburb_coordinates.sort_values('VIC_LOCA_2').reset_index()
del suburb_coordinates['index']
suburb_coordinates=suburb_coordinates.dropna()
suburb_coordinates.head()
```

```python
for index,row_sub in suburb_coordinates.iterrows():
    #print row['c1'], row['c2']
    suburb_name='Suburb_dfs/'+suburb_coordinates['VIC_LOCA_2'].iloc[index]+'.csv'
    #os.system("clear")
    print(suburb_name)
    column_names = ["VIC_LOCA_2", "Long", "Lat"]
    df_temp = pd.DataFrame(columns = column_names)
    
    df_temp=df_plants[df_plants['VIC_LOCA_2']==row_sub['VIC_LOCA_2']]
    
    #for i,row_full in df_plants.iterrows():
        #if row_sub['VIC_LOCA_2']==row_full['VIC_LOCA_2']:
            #df_temp.loc[df_plants.index[i]] = df_plants.iloc[i]
    df_temp.to_csv(suburb_name, index = False)
    
```
